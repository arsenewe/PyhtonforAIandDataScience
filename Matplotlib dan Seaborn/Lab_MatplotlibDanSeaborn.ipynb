```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
xPoints = np.array([7,20,34])
yPoints = np.array([23, 45, 67])

plt.plot(xPoints,yPoints)
plt.show()
```

```python
plt.plot(xPoints,yPoints,marker='o')
plt.show()
```

```python
xPoints = np.array([7,20,34])
yPoints = np.array([23, 4, 67])

plt.plot(xPoints,yPoints)
plt.show()
```

```python
xPoints = np.array([7,20,34])
yPoints = np.array([23, 4, 67])

plt.plot(xPoints,yPoints, ':')
plt.show()
```

```python
xPoints = np.array([7,20,34])
yPoints = np.array([23, 4, 67])

plt.plot(xPoints,yPoints,marker='o',linestyle=':',c='r')
plt.show()
```

```python
xPoints = np.array([7,20,34])
yPoints = np.array([23, 4, 67])

plt.plot(xPoints,yPoints,'o:r')
plt.show()
```

```python
xPoints = np.array([7,20,34])
yPoints = np.array([23, 4, 67])

plt.plot(xPoints,yPoints,marker='o',ms='12',mec='red', mfc='#4CAF50')
plt.show()
```

```python
xPoints = np.array([7,20,34])
yPoints = np.array([23, 4, 67])

plt.plot(xPoints,yPoints,marker='o',ms='12',mec='red', mfc='#4CAF50')
plt.title("Chart Ngawur")
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.show()
```

```python
font_title = {'family': 'serif', 'color': 'blue', 'size': 20}
font_label = {'family': 'serif', 'color': 'darkred', 'size': 12}

xPoints = np.array([7,20,34])
yPoints = np.array([23, 4, 67])

plt.plot(xPoints,yPoints,marker='o',ms='12',mec='red', mfc='#4CAF50')
plt.title("Chart Ngawur", fontdict=font_title)
plt.xlabel("Sumbu X", fontdict=font_label)
plt.ylabel("Sumbu Y", fontdict=font_label)
plt.show()
```

```python
xPoints = np.array([7,20,34])
yPoints = np.array([23, 4, 67])

plt.plot(xPoints,yPoints,marker='o',ms='12',mec='red', mfc='#4CAF50')
plt.title("Chart Ngawur")
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.grid(axis='y', color='green', linestyle='--', linewidth=0.5)
plt.show()
```
